```python
import re

cpu_usage = []
memory_usage = []
gpu_usage = []
gpu_memory_usage = []
gpu_power_usage = []

patterns = {
    'CPU Usage': (re.compile(r'CPU Usage: (\d+.\d+)%'), cpu_usage),
    'Memory Usage': (re.compile(r': Memory Usage: (\d+.\d+)%'), memory_usage),
    'GPU 0 Usage': (re.compile(r'GPU 0 Usage: (\d+\.?\d*)%'), gpu_usage),
    'GPU 0 Memory Usage': (re.compile(r'GPU 0 Memory Usage: (\d+.\d+)%'), gpu_memory_usage),
    'GPU 0 Power Usage': (re.compile(r'GPU 0 Power Usage: (\d+.\d+)%'), gpu_power_usage),
}

with open('logs/2023-05-27.log', 'r') as f:
    lines = f.readlines()

for line in lines:
    for usage_type, (pattern, usage_list) in patterns.items():
        match = pattern.search(line)
        if match:
            usage = float(match.group(1))
            usage_list.append(usage)

print('CPU Usage:', cpu_usage)
print('Memory Usage:', memory_usage)
print('GPU Usage:', gpu_usage)
print('GPU Memory Usage:', gpu_memory_usage)
print('GPU Power Usage:', gpu_power_usage)
```

```python
num = 5.1

memory_usage = [max(0, x - num) for x in memory_usage]

print(memory_usage)
```

```python
num = 5.1

gpu_power_usage = [max(0, x*170) for x in gpu_power_usage]

print(gpu_power_usage)
```

```python
num = 5.1

gpu_memory_usage = [max(0, x/100*12) for x in gpu_memory_usage]

print(gpu_memory_usage)
```

```python
import matplotlib.pyplot as plt


x = [i for i in range(len(gpu_memory_usage))]
y = gpu_power_usage

plt.figure(figsize=(10, 5))
# plot lines
plt.plot(x, y)
# , label='list1'
plt.xlabel('Time')
plt.ylabel('GPU Power(Watts)')

plt.title('GPU Power Usage')

plt.savefig('GPU Power Usage.png')
plt.legend()

plt.show()

plt.close()
```
